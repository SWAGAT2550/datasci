```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/homeprices.csv")
df
```

```python
#Using pandas to create dummy variables
dummie = pd.get_dummies(df.town)
dummie
```

```python
merged = pd.concat([df,dummie],axis='columns')
merged
```

```python
final = merged.drop(['town'], axis='columns')
final
```

```python
final = final.drop(['west windsor'], axis='columns')
final 
```

```python
X = final.drop('price', axis='columns')
X
```

```python
y = final.price
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.predict(X)
```

```python
model.score(X,y)
```

```python
model.predict([[3400,0,0]])
```

```python
df1 = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/Exercise/carprices.csv")
df1
```

```python
dummies1= pd.get_dummies(df1["Car Model"])
dummies1
```

```python
merged = pd.concat([df1,dummies1],axis='columns')
merged
```

```python
final = merged.drop(["Car Model","Mercedez Benz C class"],axis='columns')
final
```

```python
X = final.drop('Sell Price($)',axis='columns')
X
```

```python
Y = final['Sell Price($)']
Y
```

```python
model1 = LinearRegression()

model1.fit(X,Y)
```

```python
model1.score(X,Y)
```

```python
model1.predict([[45000,4,0,1]])
```
